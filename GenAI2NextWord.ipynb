```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()
```

```python
def predict_next_word(text, top_k=5):
    input_ids = tokenizer.encode(text, return_tensors="pt")

    with torch.no_grad():
        outputs = model(input_ids)
        logits = outputs.logits

    last_token_logits = logits[0, -1, :]

    top_k_tokens = torch.topk(last_token_logits, top_k).indices

    predictions = [tokenizer.decode([token]).strip() for token in top_k_tokens]
    return predictions
```

```python
input_text = "India is my"
next_words = predict_next_word(input_text)

print("Input text:", input_text)
print("Predicted next words:")
for word in next_words:
    print("-", word)
```
