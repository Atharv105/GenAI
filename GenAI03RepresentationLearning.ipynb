```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
print("Original Data Shape: ",X.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
means = np.mean(X_scaled,axis=0)
variances = np.var(X_scaled,axis=0)

print("\nLearned Feature Representations:")
for i in range (len(means)):
  print(f"Feature {i+1}: Mean = {means[i]:.4f}, Variance = {variances[i]:.4f}")
```

```python
num_samples = 100
generated_data = np.random.normal(means, np.sqrt(variances), size=(num_samples, X.shape[1]))
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis')
plt.title('Original Data (First Two Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(generated_data[:,0],generated_data[:,1], color='red')
plt.title('Generated Data (First Two Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
